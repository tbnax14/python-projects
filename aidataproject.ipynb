```python
# prompt: limit the amount of words shown in the description to 20

import pandas as pd
from datetime import datetime

df = pd.read_csv('/content/postings.csv')

# Convert the 'listed_time' column to datetime objects
df['listed_time'] = pd.to_datetime(df['listed_time'], unit='ms', origin='unix')

# Convert the 'expiry' column to datetime object
df['expiry'] = pd.to_datetime(df['expiry'], unit='ms', origin='unix')

# Format the datetime objects to a readable format (e.g., YYYY-MM-DD)
df['listed_time'] = df['listed_time'].dt.strftime('%Y-%m-%d')

# Format the datetime objects to a readable format (e.g., YYYY-MM-DD)
df['expiry'] = df['expiry'].dt.strftime('%Y-%m-%d')

# Limit the description to 20 words
df['description'] = df['description'].str.split().str[:50].str.join(' ')
columns_to_display = ['company_name', 'title', 'description', 'location','expiry','listed_time']  # Replace with actual column names
df[columns_to_display].head()
```

```python
# prompt: display postings in which the description or title has the word 'software', 'technology' or 'computer' in it

keywords = ['software', 'technology', 'computer']

# Create a boolean mask for rows where the description or title contains any of the keywords
mask = df['description'].str.contains('|'.join(keywords), case=False) | df['title'].str.contains('|'.join(keywords), case=False)

# Filter the DataFrame using the mask
filtered_df = df[mask]

# Display the filtered DataFrame
columns_to_display = ['company_name', 'title', 'description', 'max_salary', 'views', 'location', 'listed_time', 'expiry']
filtered_df[columns_to_display]
```

```python
# prompt: can you convert the max_salary column to a readable USd amount for each posting

# Assuming 'max_salary' column contains numerical values representing salary in USD
def format_salary(salary):
  """Formats a salary value into a readable USD amount."""
  if pd.isnull(salary):
    return "Not specified"
  return "${:,.2f}".format(salary)  # Formats as USD with two decimal places

# Apply the format_salary function to the 'max_salary' column
filtered_df['max_salary_formatted'] = filtered_df['max_salary'].apply(format_salary)

# Display the filtered DataFrame with the formatted salary
columns_to_display = ['company_name', 'title', 'description', 'max_salary_formatted', 'views', 'location', 'listed_time', 'expiry']
filtered_df[columns_to_display]
```
